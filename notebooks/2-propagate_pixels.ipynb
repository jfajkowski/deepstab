```python
from deepstab.flow import estimate_flow, warp_tensor
from deepstab.utils import cutout_mask
from deepstab.visualize import animate_sequence
%matplotlib inline
```

```python
from matplotlib import  rc
rc('animation', html='html5')

from torchvision import transforms
from deepstab.load import ObjectMaskVideoDataset
from deepstab.pwcnet import PWCNet
```

```python
frame_dir = '../data/raw/DAVIS/JPEGImages/480p/breakdance'
mask_dir = '../data/raw/DAVIS/Annotations_unsupervised/480p/breakdance'
model_path = '../models/base_pwc/pwc_net.pth.tar'
```

```python
plain_transforms = transforms.Compose([
    transforms.CenterCrop((512, 512))
])
further_transforms = transforms.Compose([
    transforms.ToTensor(),
    transforms.Lambda(lambda x: x.float().flip(2) * (1.0 / 255.0))
])
plain_dataset = ObjectMaskVideoDataset([frame_dir], [mask_dir], transforms=plain_transforms)
sequence = plain_dataset[0]
sequence['masked_frames'] = [further_transforms(cutout_mask(frame, mask)) for frame, mask in zip(sequence['frames'], sequence['masks'])]
```

```python
model = PWCNet('../models/base_pwcnet/network-default.pytorch').cuda().eval()

result_masked_frames = []
result_warped_frames = []
for i in range(len(sequence['masked_frames']) - 1):
    flow = estimate_flow(model, sequence['masked_frames'][i], sequence['masked_frames'][i + 1]).detach().flip(2)
    warped_frame = warp_tensor(sequence['masked_frames'][i], flow)
    result_masked_frames.append(sequence['masked_frames'][i + 1].flip(2).permute(1, 2, 0) * 255.0)
    result_warped_frames.append(warped_frame.flip(2).permute(1, 2, 0) * 255.0)
animate_sequence(result_masked_frames, result_warped_frames)
```
