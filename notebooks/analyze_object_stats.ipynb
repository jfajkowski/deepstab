```python
%matplotlib inline

import glob
import pandas as pd
import matplotlib.pyplot as plt

from os.path import basename, dirname

from inpainting.load import load_dataframe
```

```python
data_dir = '../data'
```

```python
object_stats_dfs = []
for path in glob.glob(f'{data_dir}/interim/DAVIS/ObjectStats/*/aggregated_object_stats.csv'):
    object_stats_df = load_dataframe(path)
    if object_stats_df is not None:
        object_stats_df['object_sequence_name'] = basename(dirname(path))
        object_stats_dfs.append(object_stats_df)
object_stats_df = pd.concat(object_stats_dfs)

fig = plt.figure(figsize = (10, 5))
ax = fig.gca()

object_stats_df = object_stats_df[['min_size', 'mean_size', 'max_size', 'presence']]
print(f'Number of objects: {len(object_stats_df)}')
plots = object_stats_df.hist(bins=50, ax=ax)
plots[0][0].title.set_text('Minimalny rozmiar [% powierzchni klatki]')
plots[0][1].title.set_text('Maksymalny rozmiar [% powierzchni klatki]')
plots[1][0].title.set_text('Średni rozmiar [% powierzchni klatki]')
plots[1][1].title.set_text('Obecność [% liczby klatek w sekwencji]')
```
